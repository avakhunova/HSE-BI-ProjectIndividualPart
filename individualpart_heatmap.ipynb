```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/log.csv')
```

```python
df
```

```python

df = df.rename(index={0: 'SUV39H1', 1: 'H2A', 2: 'H2B', 3: 'H3', 4: 'H4'})
df
```

```python
df = df.drop ('Unnamed: 0', axis = 1)
df
```

```python
df = df.replace(',', '.', regex=True)
```

```python
df = df.astype(float)
```

```python
f, ax = plt.subplots(figsize=(20,5))
colors = ['#000000','#1c001a','#290026','#40003b','#52024b','#b700cf','#ffffff']
ax = sns.heatmap(df, cmap = colors, annot = df)
ax.set_yticklabels(df.index.values)
plt.show()
```
